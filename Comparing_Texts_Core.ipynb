### Imports, Functions

```python
# Imports
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.probability import FreqDist
from wordcloud import WordCloud
pd.set_option('display.max_colwidth', 300)
```

```python
# Custom functions
def batch_preprocess_texts(
    texts,
    nlp=None,
    remove_stopwords=True,
    remove_punct=True,
    use_lemmas=False,
    disable=["ner"],
    batch_size=50,
    n_process=-1,
):
    """Efficiently preprocess a collection of texts using nlp.pipe()

    Args:
        texts (collection of strings): collection of texts to process (e.g. df['text'])
        nlp (spacy pipe), optional): Spacy nlp pipe. Defaults to None; if None, it creates a default 'en_core_web_sm' pipe.
        remove_stopwords (bool, optional): Controls stopword removal. Defaults to True.
        remove_punct (bool, optional): Controls punctuation removal. Defaults to True.
        use_lemmas (bool, optional): lemmatize tokens. Defaults to False.
        disable (list of strings, optional): named pipeline elements to disable. Defaults to ["ner"]: Used with nlp.pipe(disable=disable)
        batch_size (int, optional): Number of texts to process in a batch. Defaults to 50.
        n_process (int, optional): Number of CPU processors to use. Defaults to -1 (meaning all CPU cores).

    Returns:
        list of tokens
    """
    # from tqdm.notebook import tqdm
    from tqdm import tqdm

    if nlp is None:
        nlp = spacy.load("en_core_web_sm")

    processed_texts = []

    for doc in tqdm(nlp.pipe(texts, disable=disable, batch_size=batch_size, n_process=n_process)):
        tokens = []
        for token in doc:
            # Check if should remove stopwords and if token is stopword
            if (remove_stopwords == True) and (token.is_stop == True):
                # Continue the loop with the next token
                continue

            # Check if should remove stopwords and if token is stopword
            if (remove_punct == True) and (token.is_punct == True):
                continue

            # Check if should remove stopwords and if token is stopword
            if (remove_punct == True) and (token.is_space == True):
                continue

            
            ## Determine final form of output list of tokens/lemmas
            if use_lemmas:
                tokens.append(token.lemma_.lower())
            else:
                tokens.append(token.text.lower())

        processed_texts.append(tokens)
    return processed_texts
```

```python
# Load data
df = pd.read_csv("Data/Fake_Real_News_Data.csv")
df.head()
```

### 1. Clean the Data

```python
# Drop unneeded col
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Drop duplicates
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
# Check for missing values
df.isna().sum()
```

### 2. Prepare the data

- Create 3 new columns: Tokenized text, Lemmatized text, and Joined Lemmatized text.

**1. Tokenized - Keep stopwords and punctuation**

```python
# Set text to lower
df['text_lower'] = df['text'].str.lower()
df.head()
```

```python
# Get tokenized text
df['tokenized'] = df['text_lower'].str.split()
df.head()
```

**2. Lemmatized text**

```python
# Create nlp pipeline with parser and object recognizer disabled
import spacy
nlp_light = spacy.load("en_core_web_sm", disable=['parser', 'ner'])
nlp_light.pipe_names
```

```python
# Add column of lemmatized text
df['lemmatized'] = batch_preprocess_texts(df['text'], nlp=nlp_light, use_lemmas=True)
df.head()
```

**3. Joined Lemmatized Text**

```python
# Add column of single strings containing all words
df['lemmatized_joined'] = df['lemmatized'].map(lambda x: " ".join(x))
df.head()
```

### 4. Compare Word Frequencies

- Create and plot the frequency distribution plot for the 20 most common words in Real and Fake articles (2 plots total). *Use lemmatized text.*
- Create word clouds for Real and Fake articles (2 word clouds total). *Use lemmatized joined text.*

#### Create Frequency Plots

```python
# Get separate dataframes for Real and Fake articles
# REAL articles
real_df = df[df['label']=='REAL']
real_df.head(2)
```

```python
# FAKE articles
fake_df = df[df['label']=='FAKE']
fake_df.head(2)
```

```python
## Get frequency of tokens
# Create list of all tokens in all documents
real_lemmas = real_df['lemmatized'].explode().to_list()
fake_lemmas = fake_df['lemmatized'].explode().to_list()
```

```python
# Get frequencies
real_freq_dist = FreqDist(real_lemmas)
fake_freq_dist = FreqDist(fake_lemmas)
```

```python
# Plot frequencies
real_freq_dist.plot(20, title='Frequency of Words in Real Articles')
fake_freq_dist.plot(20, title='Frequency of Words in Fake Articles');
```

#### Create Word Clouds

```python
# Define a custom function for plotting the Real and Fake WordClouds
def plot_wordclouds(real_cloud, fake_cloud, title='Comparing Word Usage'):
    """Plots the wordlcouds for our two groups"""
    ## Plot the Images
    fig, axes = plt.subplots(ncols=2, figsize=(10, 5))
    axes[0].imshow(real_cloud)
    axes[0].set_title('Real Articles')
    axes[0].axis('off')
    
    axes[1].imshow(fake_cloud)
    axes[1].set_title('Fake Ratings')
    axes[1].axis('off')
    fig.tight_layout()
    
    fig.suptitle(title,y=1.0, fontsize='xx-large');
    return fig
```

```python
# Get a list of all lemmatized text in all articles
real_cloud_list = " ".join(real_df['lemmatized_joined'])
fake_cloud_list = " ".join(fake_df['lemmatized_joined'])
```

```python
# Generate the WordCloud images
wordcloud_kws = dict(min_word_length=2, width=800, height=600, random_state=42)
real_cloud = WordCloud(**wordcloud_kws, colormap='Greens').generate(real_cloud_list)
fake_cloud = WordCloud(**wordcloud_kws, colormap='Reds').generate(fake_cloud_list)
```

```python
# Display the figure
fig = plot_wordclouds(real_cloud, fake_cloud, title='Comparing Real and Fake News Articles')
# Saving a copy for README
fig.savefig('images/wordclouds-real-fake.png', dpi=300, bbox_inches='tight', transparent=False)
```
